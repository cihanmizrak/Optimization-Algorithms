```python
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Randmo Walk
def random_walk_3d(steps):
    x, y, z = 0, 0, 0
    points = [(x, y, z)]

    for _ in range(steps):
        direction = random.choice(["up", "down", "left", "right", "forward", "backward"])
        if direction == "up":
            y += 1
        elif direction == "down":
            y -= 1
        elif direction == "left":
            x -= 1
        elif direction == "right":
            x += 1
        elif direction == "forward":
            z += 1
        else:
            z -= 1
        points.append((x, y, z))

    return points

# Create points with random walks
num_steps = 50
points = random_walk_3d(num_steps)

# Draw 3D graph
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

x_coords, y_coords, z_coords = zip(*points)
ax.plot(x_coords, y_coords, z_coords, marker='o')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("3D Graphic of Random Walk Algorithm")
plt.show()
```

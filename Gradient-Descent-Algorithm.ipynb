```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the quadratic function to minimize (example function)
def quadratic_function(x, y):
    return x**2 + y**2

# Gradient of the quadratic function
def gradient(x, y):
    grad_x = 2 * x
    grad_y = 2 * y
    return np.array([grad_x, grad_y])

# Gradient Descent Algorithm
def gradient_descent(initial_point, learning_rate, num_iterations):
    trajectory = [initial_point]
    current_point = initial_point.copy()
    
    for _ in range(num_iterations):
        grad = gradient(*current_point)
        current_point = current_point - learning_rate * grad
        trajectory.append(current_point)
    
    return np.array(trajectory)

# Parameters for gradient descent
initial_point = np.array([3.0, 4.0])
learning_rate = 0.3
num_iterations = 20

# Run gradient descent
trajectory = gradient_descent(initial_point, learning_rate, num_iterations)

# Create a 3D plot
x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = quadratic_function(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the function surface
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.5)

# Plot the trajectory of gradient descent
ax.plot(trajectory[:, 0], trajectory[:, 1], quadratic_function(*trajectory.T), marker='o', color='r')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Function Value')

plt.show()
```
